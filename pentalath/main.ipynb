```python
def cube_to_axial(cube):
    q = cube[:, 0]
    r = cube[:, 2]
    return (q, r)

def axial_to_cube(hexchoord):
    z = hexchoord[0]
    x = hexchoord[1]
    y = -x-z
    return (x, y, z)
```

```python
def pos_occupied(x, y, z):
    global piece_position
    if (x, y, z) in piece_position:
        return True
    return False
    
def make_move(to_position):
    global piece_position
    global move_history
    global move_counter
    global side
    piece_position[to_position] = side
    move_history[move_counter, to_position] = [ side]
    move_counter += 1
    
def undo_make_move(to_position):
    global piece_position
    global move_history
    global move_counter
    global side
    del piece_position[to_position]
#     print(move_history)
    move_counter -= 1
    del move_history[move_counter, to_position]
    
def reset():
    global piece_position
    piece_position = {}
    
    global move_history
    move_history = {}
    
    global move_counter
    move_counter = 0
    
    global hex_grid
    global hex_grid_temp
    
    hex_grid = hex_grid_temp
    
    global side
    side = 'B'      
            
            
def horizontal_win(x, y, z):
    if (x + 4) > (6 - z):
        return False
#     print('hi1') #(-1, -2, 3) --> -1, 0, 1, 2, 3 == -2, -3, -4, -5, -6 == 2
    if pos_occupied(x, y, z) and \
        pos_occupied(x+1, y-1, z) and \
        pos_occupied(x+2, y-2, z) and \
        pos_occupied(x+3, y-3, z) and \
        pos_occupied(x+4, y-4, z) and \
        (piece_position[(x, y, z)] == \
        piece_position[(x+1, y-1, z)] == \
        piece_position[(x+2, y-2, z)] == \
        piece_position[(x+3, y-3, z)] == \
        piece_position[(x+4, y-4, z)]):
        return True, (x, y, z), (x+1, y-1, z), (x+2, y-2, z), (x+3, y-3, z), (x+4, y-4, z)
    return False
    
def vertical_win_left(x, y, z):
    if  (z < 4 and x > 2) or  y - 4  < -(6 + x):
        return False
#     print('hi2')
    if pos_occupied(x, y, z) and \
        pos_occupied(x, y-1, z+1) and \
        pos_occupied(x, y-2, z+2) and \
        pos_occupied(x, y-3, z+3) and \
        pos_occupied(x, y-4, z+4) and \
        (piece_position[(x, y, z)] == \
        piece_position[(x, y-1, z+1)] == \
        piece_position[(x, y-2, z+2)] == \
        piece_position[(x, y-3, z+3)] == \
        piece_position[(x, y-4, z+4)]):
        return True, (x, y, z), (x, y-1, z+1), (x, y-2, z+2), (x, y-3, z+3), (x, y-4, z+4)
    return False

def vertical_win_right(x, y, z):
    if  (z < 4 and y > 2) or  x - 4  < -(6 + y):
        return False
#     print('hii')
    if pos_occupied(x, y, z) and \
        pos_occupied(x-1, y, z+1) and \
        pos_occupied(x-2, y, z+2) and \
        pos_occupied(x-3, y, z+3) and \
        pos_occupied(x-4, y, z+4) and \
        (piece_position[(x, y, z)] == \
        piece_position[(x-1, y, z+1)] == \
        piece_position[(x-2, y, z+2)] == \
        piece_position[(x-3, y, z+3)] == \
        piece_position[(x-4, y, z+4)]):
        return True, (x, y, z), (x-1, y, z+1), (x-2, y, z+2), (x-3, y, z+3), (x-4, y, z+4)
    return False
    
def check_win():
    for j in range(7): # row, z
        for i in range(-6, 7-j): # col, y, x
            x, y, z = axial_to_cube((j, i))
#             print(x, y, z)
            if horizontal_win(x, y, z):
                return horizontal_win(x, y, z)
            elif vertical_win_left(x, y, z):
                return vertical_win_left(x, y, z)
            elif vertical_win_right(x, y, z):
                return vertical_win_right(x, y, z)
            
    return False


def printBoard(board):
    row = ''
    tab = ' '
    
    """ Print board by converting axial to cube positions
    
    """
    for j in range(7): # row, z
        for i in range(-6, 7-j): # col, y, x
            x, y, z = axial_to_cube((j, i))
            
            if pos_occupied(x, y, z):
                row += piece_position[(x, y, z)]        
            else:
                row += '_'
                
            row += ' '
            
        row += '\n'
        row += tab
        tab += ' '

        
        
    print(row)
```
